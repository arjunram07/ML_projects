## 2009


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
a=pd.read_csv('xrd.csv')
b=a[['serial','INC','BJP','BSP']]
b.head()
```

```python
c=a[['INC','BJP','BSP']]
sns.catplot(kind='box',data=c)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
serial=list(b['serial'])
BSP=list(b['BSP'])
INC=list(b['INC'])
BJP=list(b['BJP'])
plt.xlabel('serial numbers of polling booths')
plt.bar(serial,INC,color='dodgerblue',label='INC')
plt.bar(serial,BJP,color='orange',label='BJP')
plt.bar(serial,BSP,color='green',label='BSP')
plt.legend()
plt.subplot(122)
plt.xlabel('serial numbers of polling booths')
plt.plot(serial,BSP,color='green',label='BSP')
plt.plot(serial,BJP,color='orange',label='BJP')
plt.plot(serial,INC,color='dodgerblue',label='INC')
plt.legend()
plt.show()
```

##### From the above plots we can understand that the INC are the majority in the 2009 elections.
